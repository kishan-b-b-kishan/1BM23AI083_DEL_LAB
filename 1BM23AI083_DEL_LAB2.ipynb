```python
#2.2
import numpy as np
def sigmoid(x):
    return 1/(1+np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0,x)

class neuron:
    def __init__(self,weights,bias,activation='sigmoid'):
        self.weights=weights
        self.bias=bias
        self.activation=activation
    def fwd(self,ip):
        x=np.dot(ip,self.weights)+self.bias
        
        if self.activation=='sigmoid':
            return sigmoid(x)
        elif self.activation=='tanh':
            return tanh(x)
        elif self.activation=='relu':
            return relu(x)
        else:
            raise ValueError('unknown activation fn')

ips=np.array([0.5,-1.2,3.0])
w=[0.4,-0.6,0.2]
bias=0.1

for act in ['sigmoid','tanh','relu']:
    n=neuron(w,bias,activation=act)
    op=n.fwd(ips)
    print(f'{act}\nouput = {op:4f}\n')



x = np.linspace(-5, 5, 400) #inputs

y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_relu = relu(x)

fig, axs = plt.subplots(1, 3, figsize=(15, 4))

#sigmoid
axs[0].plot(x, y_sigmoid, color='blue')
axs[0].set_title('Sigmoid Activation')
axs[0].grid(True)

#tanh
axs[1].plot(x, y_tanh, color='green')
axs[1].set_title('Tanh Activation')
axs[1].grid(True)

#relu
axs[2].plot(x, y_relu, color='red')
axs[2].set_title('ReLU Activation')
axs[2].grid(True)

plt.tight_layout()
plt.show()

```
